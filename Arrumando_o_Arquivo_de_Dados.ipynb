## Removing Null and None  and shapping to numeric

```python
profiles['current_salary'] = profiles['current_salary'].str.replace(',', '')
profiles['current_salary'] = pd.to_numeric(profiles['current_salary'])
```

```python
feet = pd.to_numeric(profiles['height'].str[0])
inch = pd.to_numeric(profiles['height'].str[2])
inch = inch*0.83333
height = feet+inch
profiles['height'] = height.round(2)
```

## Arrumando a data ad Objeto para Datetime

```python
profiles['birth_date'] = pd.to_datetime(profiles['birth_date'])
profiles['death_date'] = pd.to_datetime(profiles['death_date'], errors = 'ignore')
```
